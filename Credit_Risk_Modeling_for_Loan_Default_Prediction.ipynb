```python
# Phase 2 ‚Äî Data Preprocessing & Feature Engineering
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# ----------------------------
# 1. Load EDA output
# ----------------------------
df = pd.read_csv("/content/data/phase1_eda_output.csv")
print(f"‚úÖ Loaded dataset with shape: {df.shape}")

# ----------------------------
# 2. Keep relevant columns for PD modeling
# ----------------------------
# Drop columns that leak post-loan info or identifiers
leakage_cols = ['loan_status', 'title', 'address']
df = df.drop(columns=[c for c in leakage_cols if c in df.columns])

# ----------------------------
# 3. Handle data types and numeric conversions
# ----------------------------
# Convert int_rate and revol_util if needed
if df['int_rate'].max() > 1:  # sometimes in %
    df['int_rate'] = df['int_rate'] / 100
if df['revol_util'].max() > 1:
    df['revol_util'] = df['revol_util'] / 100

# ----------------------------
# 4. Feature engineering ‚Äî new derived features
# ----------------------------
if {'loan_amnt','annual_inc'}.issubset(df.columns):
    df['loan_to_income'] = df['loan_amnt'] / (df['annual_inc'] + 1)

if {'dti','revol_util'}.issubset(df.columns):
    df['credit_stress'] = df['dti'] * df['revol_util']

if 'emp_length' in df.columns:
    df['emp_length'] = df['emp_length'].fillna("0 years")
    df['emp_length_num'] = (
        df['emp_length']
        .str.extract(r'(\d+)')
        .astype(float)
        .fillna(0)
    )

# ----------------------------
# 5. Select features
# ----------------------------
numeric_features = [
    'loan_amnt','int_rate','installment','annual_inc','dti',
    'open_acc','pub_rec','revol_bal','revol_util','total_acc',
    'mort_acc','loan_to_income','credit_stress','emp_length_num'
]
numeric_features = [f for f in numeric_features if f in df.columns]

categorical_features = ['grade','home_ownership','purpose','verification_status','application_type']
categorical_features = [f for f in categorical_features if f in df.columns]

target = 'default_flag'

print(f"Numeric features: {len(numeric_features)} ‚Äî {numeric_features}")
print(f"Categorical features: {len(categorical_features)} ‚Äî {categorical_features}")

# ----------------------------
# 6. Preprocessing pipelines
# ----------------------------
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# ----------------------------
# 7. Train‚Äìtest split
# ----------------------------
X = df[numeric_features + categorical_features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.3, random_state=42
)
print(f"\n‚úÖ Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ----------------------------
# 8. Fit preprocessing on training set and transform
# ----------------------------
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print(f"Preprocessed feature matrix shape: {X_train_processed.shape}")

# ----------------------------
# 9. Save preprocessing pipeline
# ----------------------------
import joblib
joblib.dump(preprocessor, "/content/data/pd_preprocessor.pkl")
print("üìÅ Saved preprocessing pipeline ‚Üí /mnt/data/pd_preprocessor.pkl")

# ----------------------------
# 10. Optional ‚Äî check a few transformed features
# ----------------------------
print("\nSample of preprocessed data (first 5 rows):")
print(pd.DataFrame(X_train_processed).head())
```

```python
# Phase 3 ‚Äî Modeling: PD (Logistic Regression baseline + LightGBM)
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, classification_report, brier_score_loss
from sklearn.model_selection import train_test_split
import lightgbm as lgb
import seaborn as sns

SEED = 42

# 0. Load preprocessor and data
preproc = joblib.load("/content/data/pd_preprocessor.pkl")
df = pd.read_csv("/content/data/phase1_eda_output.csv", low_memory=False)
target = "default_flag"

# ----------------------------
# Re-create Feature Engineering from Phase 2
# ----------------------------
# Convert int_rate and revol_util if needed (copied from Phase 2)
if df['int_rate'].max() > 1:
    df['int_rate'] = df['int_rate'] / 100
if df['revol_util'].max() > 1:
    df['revol_util'] = df['revol_util'] / 100

# Feature engineering ‚Äî new derived features (copied from Phase 2)
if {'loan_amnt','annual_inc'}.issubset(df.columns):
    df['loan_to_income'] = df['loan_amnt'] / (df['annual_inc'] + 1)

if {'dti','revol_util'}.issubset(df.columns):
    df['credit_stress'] = df['dti'] * df['revol_util']

if 'emp_length' in df.columns:
    df['emp_length'] = df['emp_length'].fillna("0 years")
    df['emp_length_num'] = (
        df['emp_length']
        .str.extract(r'(\d+)')
        .astype(float)
        .fillna(0)
    )


# Recreate feature lists used in Phase 2 (must match preprocessor columns)
numeric_features = [
    'loan_amnt','int_rate','installment','annual_inc','dti',
    'open_acc','pub_rec','revol_bal','revol_util','total_acc',
    'mort_acc','loan_to_income','credit_stress','emp_length_num'
]
categorical_features = ['grade','home_ownership','purpose','verification_status','application_type']
# Include the engineered features in the features list
features = [f for f in numeric_features + categorical_features if f in df.columns]


# Prepare X,y and split (use the same split strategy as Phase 2)
X = df[features].copy()
y = df[target].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=SEED)

# Transform using saved preprocessor
X_train_proc = preproc.transform(X_train)
X_test_proc = preproc.transform(X_test)

# ----------------------------
# Utility functions
# ----------------------------
def ks_stat(y_true, y_score):
    """Compute KS statistic and return max difference."""
    fpr, tpr, thr = roc_curve(y_true, y_score)
    ks = np.max(np.abs(tpr - fpr))
    return ks

def decile_table(y_true, y_score, n_bins=10):
    dfb = pd.DataFrame({"y": y_true, "score": y_score})
    dfb["decile"] = pd.qcut(dfb["score"].rank(method="first"), q=n_bins, labels=False) + 1
    agg = dfb.groupby("decile").agg(
        count=("y","size"),
        defaults=("y","sum"),
        avg_score=("score","mean")
    ).sort_index(ascending=False)  # highest scores first
    agg["default_rate"] = agg["defaults"] / agg["count"]
    agg["cum_defaults"] = agg["defaults"].cumsum()
    agg["cum_default_pct"] = agg["cum_defaults"] / agg["defaults"].sum()
    return agg

def plot_calibration(y_true, y_prob, n_bins=10, title="Calibration plot"):
    prob_true, prob_pred = calibration_curve(y_true, y_prob, n_bins=n_bins, strategy='quantile')
    plt.figure(figsize=(6,5))
    plt.plot(prob_pred, prob_true, marker='o', label='Calibration')
    plt.plot([0,1],[0,1],'--', color='grey', label='Perfect')
    plt.xlabel('Mean predicted probability')
    plt.ylabel('Fraction of positives')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# ----------------------------
# 1) Baseline: Logistic Regression (with simple L2)
# ----------------------------
print("Training logistic regression baseline...")
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=SEED)
lr.fit(X_train_proc, y_train)

# Calibrate logistic (sigmoid)
cal_lr = CalibratedClassifierCV(lr, method='sigmoid', cv='prefit')
cal_lr.fit(X_train_proc, y_train)

lr_proba = cal_lr.predict_proba(X_test_proc)[:,1]
lr_auc = roc_auc_score(y_test, lr_proba)
lr_ks = ks_stat(y_test, lr_proba)
lr_brier = brier_score_loss(y_test, lr_proba)

print("\nLogistic Regression (calibrated) metrics:")
print("AUC:", round(lr_auc,4))
print("KS:", round(lr_ks,4))
print("Brier:", round(lr_brier,4))

# Decile table
print("\nTop deciles (LR):")
print(decile_table(y_test, lr_proba).head())

# Calibration plot
plot_calibration(y_test, lr_proba, n_bins=10, title="Calibration (Logistic Regression)")

# ----------------------------
# 2) LightGBM model (preferred)
# ----------------------------
print("\nTraining LightGBM model...")
# Use dataset with preprocessor output; LightGBM can accept numpy arrays directly
train_data = lgb.Dataset(X_train_proc, label=y_train)
valid_data = lgb.Dataset(X_test_proc, label=y_test, reference=train_data)

params = {
    "objective": "binary",
    "metric": "auc",
    "learning_rate": 0.05,
    "num_leaves": 64,
    "min_data_in_leaf": 100,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "seed": SEED,
    "verbosity": -1
}

bst = lgb.train(params, train_data, num_boost_round=1000, valid_sets=[valid_data],
                callbacks=[lgb.early_stopping(stopping_rounds=50), lgb.log_evaluation(period=50)]) # FIX: Changed early_stopping_rounds to callbacks and added lgb.log_evaluation

# Predict probabilities
lgb_proba = bst.predict(X_test_proc, num_iteration=bst.best_iteration)
lgb_auc = roc_auc_score(y_test, lgb_proba)
lgb_ks = ks_stat(y_test, lgb_proba)
lgb_brier = brier_score_loss(y_test, lgb_proba)

print("\nLightGBM metrics:")
print("AUC:", round(lgb_auc,4))
print("KS:", round(lgb_ks,4))
print("Brier:", round(lgb_brier,4))

# Decile table
print("\nTop deciles (LightGBM):")
print(decile_table(y_test, lgb_proba).head())

# Calibration plot
plot_calibration(y_test, lgb_proba, n_bins=10, title="Calibration (LightGBM)")

# ----------------------------
# 3) Optional: calibrate LightGBM with isotonic (if needed)
# ----------------------------
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
# NOTE: scikit-learn's CalibratedClassifierCV expects an sklearn estimator with fit/predict_proba
# You can instead use a small sklearn wrapper or use a validation set to fit a Platt scaling (Logistic) on lgb_proba.

# Simpler approach: fit Platt scaling (logistic) on an OOB/holdout set:
from sklearn.linear_model import LogisticRegression
# Create a small calibration training set using a validation split of training data
X_tr, X_cal, y_tr, y_cal = train_test_split(X_train_proc, y_train, stratify=y_train, test_size=0.1, random_state=SEED)
# Re-train LightGBM on X_tr, then predict on X_cal to get raw scores, then fit logistic on those scores
train_data2 = lgb.Dataset(X_tr, label=y_tr)
valid_data2 = lgb.Dataset(X_cal, label=y_cal)
bst2 = lgb.train(params, train_data2, num_boost_round=bst.best_iteration, valid_sets=[valid_data2]) # FIX: Removed verbose_eval=False
cal_scores = bst2.predict(X_cal)
platt = LogisticRegression(max_iter=1000)
platt.fit(cal_scores.reshape(-1,1), y_cal)
# Apply Platt to test predictions
lgb_platt_proba = platt.predict_proba(lgb_proba.reshape(-1,1))[:,1]
print("\nLightGBM + Platt AUC:", round(roc_auc_score(y_test, lgb_platt_proba),4))
print("KS (platt):", round(ks_stat(y_test, lgb_platt_proba),4))

# ----------------------------
# 4) Save artifacts (preprocessor already saved)
# ----------------------------
joblib.dump(lr, "/content/data/pd_lr_raw.pkl")
joblib.dump(cal_lr, "/content/data/pd_lr_calibrated.pkl")
bst.save_model("/content/data/pd_lightgbm.txt")
joblib.dump(platt, "/content/data/pd_lgb_platt_scaler.pkl")
print("\nSaved models to /content/data/: pd_lr_raw.pkl, pd_lr_calibrated.pkl, pd_lightgbm.txt, pd_lgb_platt_scaler.pkl")

# ----------------------------
# 5) Feature importance (approx) ‚Äî for LightGBM using preprocessor feature names
# ----------------------------
# Recover feature names from preprocessor: numeric + onehot feature names
num_features = numeric_features
# retrieve onehot names
ohe_feature_names = []
try:
    # preproc is ColumnTransformer with a named transformer 'cat' -> OneHotEncoder
    cat_transformer = preproc.named_transformers_['cat']
    ohe = cat_transformer.named_steps['encoder']
    ohe_feature_names = list(ohe.get_feature_names_out(categorical_features))
except Exception as e:
    print("Could not extract one-hot names:", e)
feature_names = num_features + ohe_feature_names

# LightGBM feature importance
imp = bst.feature_importance(importance_type='gain')
imp_df = pd.DataFrame({"feature": feature_names, "importance": imp})
imp_df = imp_df.sort_values("importance", ascending=False).head(30)
print("\nTop LightGBM features by gain:")
print(imp_df)

# ----------------------------
# 6) Produce final decile CSV and save scored sample
# ----------------------------
test_df = X_test.copy().reset_index(drop=True)
test_df[target] = y_test.reset_index(drop=True)
test_df["pd_lgb"] = lgb_proba
test_df["pd_lgb_platt"] = lgb_platt_proba
# Add LR probability for comparison
test_df["pd_lr"] = lr_proba

# Save decile table and a sample scored file
dec = decile_table(test_df[target], test_df["pd_lgb"])
dec.to_csv("/content/data/pd_decile_table_lgb.csv", index=True)
test_df.head(500).to_csv("/content/data/pd_scored_sample.csv", index=False)
print("\nSaved decile table and sample scored CSV to /content/data/")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# --- Inputs ---
# Use your test set predictions (from LightGBM + Platt scaling)
# y_true = y_test
# y_pred = lgb_platt_proba

# Define thresholds you want to evaluate
thresholds = [0.25, 0.35, 0.5]

# Store results
threshold_records = []

for t in thresholds:
    preds = (y_pred >= t).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, preds).ravel()

    precision_t = precision_score(y_true, preds)
    recall_t = recall_score(y_true, preds)
    f1_t = f1_score(y_true, preds)
    defaults_flagged = preds.sum() / len(preds)

    threshold_records.append({
        "Threshold": round(t, 2),
        "Precision": round(precision_t, 3),
        "Recall": round(recall_t, 3),
        "F1_Score": round(f1_t, 3),
        "Defaults_Flagged(%)": f"{defaults_flagged*100:.2f}%",
        "True_Defaults(%)": f"{y_true.mean()*100:.2f}%",
        "Comment": "High risk" if t>=0.5 else ("Medium risk" if t>=0.3 else "Low risk")
    })

# Create DataFrame
threshold_table = pd.DataFrame(threshold_records)

print("üìä Threshold Performance Summary:\n")
print(threshold_table)

# Save to CSV and Markdown
threshold_table.to_csv("/content/data/threshold_summary.csv", index=False)

with open("/content/data/threshold_summary.md", "w") as f:
    f.write("### Credit Risk Threshold Documentation\n\n")
    f.write("Below is the precision‚Äìrecall summary for different decision thresholds used in PD segmentation.\n\n")
    f.write(threshold_table.to_markdown(index=False))
    f.write("\n\n**Interpretation:**\n")
    f.write("- High risk = likely default ‚Üí decline / manual review\n")
    f.write("- Medium risk = borderline ‚Üí additional verification\n")
    f.write("- Low risk = likely non-default ‚Üí auto-approve\n")

print("\n‚úÖ Saved threshold documentation:")
print(" - CSV ‚Üí /content/data/threshold_summary.csv")
print(" - Markdown ‚Üí /content/data/threshold_summary.md")
```

```python
# Phase C: Basel III Risk Parameter Estimation & Portfolio Simulation
import pandas as pd, numpy as np, joblib
from scipy.stats import norm
import os

# Load model and preprocessor
DATA_DIR = "/content/data"
preproc = joblib.load(os.path.join(DATA_DIR, "pd_preprocessor_phaseB.pkl"))
model = joblib.load(os.path.join(DATA_DIR, "pd_lgb_phaseB.pkl"))

# Load dataset
df = pd.read_csv("/content/lending_club_loan_two.csv", low_memory=False)
print("Loaded dataset with shape:", df.shape)

# Create target & prepare input features again
if 'loan_status' in df.columns:
    df['default_flag'] = df['loan_status'].isin(['Charged Off', 'Default']).astype(int)
    print("Created default_flag")
else:
    if 'default_flag' not in df.columns:
        print("No loan_status or default_flag available.")
        df['default_flag'] = np.nan

# Ensure issue_d parsed if present
if 'issue_d' in df.columns:
    df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y', errors='coerce')
    print("Converted issue_d to datetime.")

# --- Re-apply Feature Engineering from Phase B ---
# This ensures the features used by the preprocessor and model are present
# recreate phase2 features if missing
if {'loan_amnt','annual_inc'}.issubset(df.columns):
    df['loan_to_income'] = df['loan_amnt'] / (df['annual_inc'] + 1)
if {'dti','revol_util'}.issubset(df.columns):
    if df['revol_util'].dtype == object:
        df['revol_util'] = df['revol_util'].astype(str).str.rstrip('%').astype(float)/100
    df['credit_stress'] = df['dti'] * df['revol_util']
if 'emp_length' in df.columns and 'emp_length_num' not in df.columns:
    df['emp_length'] = df['emp_length'].fillna("0")
    df['emp_length_num'] = (
        df['emp_length']
        .astype(str).str.extract(r'(\d+)')
        .astype(float)
        .fillna(0)
    )

# additional features from Phase B
if 'earliest_cr_line' in df.columns and 'issue_d' in df.columns:
    df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%Y', errors='coerce')
    if pd.api.types.is_datetime64_any_dtype(df['issue_d']) and pd.api.types.is_datetime64_any_dtype(df['earliest_cr_line']):
         df['credit_age_years'] = (df['issue_d'] - df['earliest_cr_line']).dt.days / 365.25
    else:
         df['credit_age_years'] = np.nan

for col in ['annual_inc','loan_amnt','installment','revol_bal']:
    if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):
        df[f'log_{col}'] = np.log1p(df[col].clip(lower=0))

if {'installment','annual_inc'}.issubset(df.columns):
    if pd.api.types.is_numeric_dtype(df['installment']) and pd.api.types.is_numeric_dtype(df['annual_inc']):
        df['installment_to_income'] = df['installment'] / (df['annual_inc'] + 1)

if 'delinq_2yrs' in df.columns and not pd.api.types.is_numeric_dtype(df['delinq_2yrs']):
     df['delinq_2yrs'] = pd.to_numeric(df['delinq_2yrs'], errors='coerce').fillna(0)
elif 'delinq_2yrs' not in df.columns:
    df['delinq_2yrs'] = 0


if {'dti','revol_util'}.issubset(df.columns):
    if pd.api.types.is_numeric_dtype(df['dti']) and pd.api.types.is_numeric_dtype(df['revol_util']):
         df['dti_revol_int'] = df['dti'] * df['revol_util']
# --- End Re-apply Feature Engineering ---


# Define feature lists used in Phase B's preprocessor (must match)
# Ensure these lists match the columns used to train the preprocessor in Phase B
# Retrieve feature names from the loaded preprocessor
try:
    numeric_features = preproc.transformers_[0][2] # Numeric feature names from preprocessor
    categorical_features = preproc.transformers_[1][2] # Categorical feature names from preprocessor
    print("Retrieved feature names from preprocessor.")
except Exception as e:
    print("Could not retrieve feature names from preprocessor:", e)
    # Fallback to manually defined lists if retrieval fails (needs manual sync)
    numeric_features = [
        'loan_amnt','int_rate','installment','annual_inc','dti','open_acc','pub_rec','revol_bal',
        'revol_util','total_acc','mort_acc','loan_to_income','credit_stress','emp_length_num',
        'credit_age_years','installment_to_income','dti_revol_int','log_annual_inc','log_loan_amnt','log_installment',
        'log_revol_bal','delinq_2yrs'
    ]
    categorical_features = ['grade','home_ownership','purpose','verification_status','application_type','term']
    print("Using manually defined feature lists (ensure these match Phase B!).")


all_features = numeric_features + categorical_features
# Keep only features that exist in the dataframe after loading and engineering
X = df[all_features].copy()
print("X shape before processing:", X.shape)

# Predict PDs
# Handle potential missing values in X before transformation
X_proc = preproc.transform(X)
print("X shape after processing:", X_proc.shape)

# Check if PD_score is already in df (e.g., from a loaded scored file)
if 'PD_score' not in df.columns:
     # Predict PD using the trained model
     df['PD_score'] = model.predict_proba(X_proc)[:, 1]
     print("Predicted PD_score using the trained model.")
else:
     print("PD_score column already exists.")


# Simulate LGD (using simple rule-based or a model if trained)
# Check if LGD model was trained (i.e., recoveries column was present in original data)
lgd_model_path = os.path.join(DATA_DIR, "lgd_model.pkl")
if os.path.exists(lgd_model_path):
    try:
        lgd_pipe = joblib.load(lgd_model_path)
        # Score LGD for all loans using the trained model
        df['LGD_model'] = lgd_pipe.predict(X_proc) # Use X_proc (all features preprocessed)
        df['LGD_model'] = df['LGD_model'].clip(lower=0.0, upper=1.0)
        print("Predicted LGD_model using trained model.")
    except Exception as e:
        print("Could not predict LGD using trained model:", e)
        # Fallback if model prediction fails
        df['LGD_model'] = 0.45 # Conservative default if model fails/not applicable
        print("Using conservative default LGD_model = 0.45 as fallback.")
else:
    # Fallback if no LGD model was trained (due to missing 'recoveries')
    # Use a simple rule or a conservative average
    def estimate_lgd_fallback(row):
        if row.get('home_ownership') == 'MORTGAGE':
            return 0.4 # Example conservative values
        elif row.get('home_ownership') == 'RENT':
            return 0.7
        else:
            return 0.5
    df['LGD_model'] = df.apply(estimate_lgd_fallback, axis=1)
    print("No LGD model found; using rule-based LGD_model fallback.")

# Simulate EAD (using simple rule-based or a model if trained)
ead_model_path = os.path.join(DATA_DIR, "ead_model.pkl")
if os.path.exists(ead_model_path):
    try:
        ead_pipe = joblib.load(ead_model_path)
         # Score EAD for all loans using the trained model
        df['EAD_model'] = ead_pipe.predict(X_proc) # Use X_proc (all features preprocessed)
        # EAD cannot be negative and should not exceed loan_amnt if available
        if 'loan_amnt' in df.columns:
             df['EAD_model'] = df['EAD_model'].clip(lower=0.0, upper=df['loan_amnt'])
        else:
            df['EAD_model'] = df['EAD_model'].clip(lower=0.0)
        print("Predicted EAD_model using trained model.")
    except Exception as e:
        print("Could not predict EAD using trained model:", e)
        # Fallback if model prediction fails
        df['EAD_model'] = df.get('loan_amnt', df.get('funded_amnt', df.get('installment', 0))) # Fallback to loan_amnt, funded_amnt, or installment
        print("Using loan_amnt (or fallback) as EAD_model fallback.")
else:
    # Fallback if no EAD model was trained
    # Use outstanding balance if available, else loan amount
    if 'out_prncp' in df.columns:
        df['EAD_model'] = df['out_prncp'].fillna(df.get('loan_amnt', df.get('funded_amnt', 0)))
        print("Using outstanding_prncp (and loan_amnt fallback) as EAD_model.")
    elif 'loan_amnt' in df.columns:
        df['EAD_model'] = df['loan_amnt']
        print("Using loan_amnt as EAD_model.")
    elif 'funded_amnt' in df.columns:
        df['EAD_model'] = df['funded_amnt']
        print("Using funded_amnt as EAD_model.")
    else:
        df['EAD_model'] = df.get('installment', 0)
        print("Using installment as EAD_model fallback.")


# Expected Loss
# Only compute if all components are available
if 'PD_score' in df.columns and 'LGD_model' in df.columns and 'EAD_model' in df.columns:
     df['Expected_Loss'] = df['PD_score'] * df['LGD_model'] * df['EAD_model']
     print("Computed Expected_Loss.")
else:
     print("Could not compute Expected_Loss: Missing PD_score, LGD_model, or EAD_model.")


# Basel III RWA Calculation (only if PD, LGD_model, EAD_model are available)
if 'PD_score' in df.columns and 'LGD_model' in df.columns and 'EAD_model' in df.columns:
    def basel_rwa(pd, lgd, ead):
        # Ensure inputs are numpy arrays for vectorized operations
        pd = np.asarray(pd)
        lgd = np.asarray(lgd)
        ead = np.asarray(ead)

        # Handle potential NaNs or non-numeric inputs gracefully
        if not (np.issubdtype(pd.dtype, np.number) and np.issubdtype(lgd.dtype, np.number) and np.issubdtype(ead.dtype, np.number)):
             print("Warning: Non-numeric inputs to basel_rwa function.")
             return np.nan # Or handle as appropriate

        # Clip PD to avoid extreme values in norm.ppf
        pd = np.clip(pd, 1e-6, 1-1e-6)

        # Handle potential division by zero or issues in norm.ppf with edge cases
        with np.errstate(divide='ignore', invalid='ignore'): # Ignore warnings for numerical issues
            R = 0.12 * (1 - np.exp(-50*pd)) / (1 - np.exp(-50)) + \
                0.24 * (1 - (1 - np.exp(-50*pd)) / (1 - np.exp(-50)))
            # Ensure R is not too close to 1 for the square root
            R = np.clip(R, 0, 0.9999) # Clip R to avoid sqrt(negative) or division by zero

            # Calculate K
            term1 = (1/np.sqrt(1-R)) * norm.ppf(pd)
            term2 = np.sqrt(R/(1-R)) * norm.ppf(0.999)

            # Handle potential NaNs or infinities from norm.ppf
            K = lgd * norm.cdf(term1 + term2) - pd * lgd
            K[np.isnan(K)] = 0 # Treat NaN K as 0 or a small value if appropriate
            K = np.clip(K, 0, None) # K should not be negative

            rwa = 12.5 * K * ead

            # Handle potential NaNs or infinities in final RWA
            rwa[np.isnan(rwa)] = 0 # Treat NaN RWA as 0 or a small value

        return rwa

    df['RWA'] = basel_rwa(df['PD_score'], df['LGD_model'], df['EAD_model'])
    print("Computed RWA.")
else:
     print("Could not compute RWA: Missing PD_score, LGD_model, or EAD_model.")


# Portfolio summary
summary_cols = ['loan_amnt','PD_score','LGD_est','LGD_model','outstanding_est','EAD_model','Expected_Loss','RWA']
existing_summary_cols = [c for c in summary_cols if c in df.columns]
if existing_summary_cols:
    summary = df[existing_summary_cols].describe()
    print("\n--- Portfolio Summary Statistics ---")
    display(summary)
else:
    print("\nNo columns available for portfolio summary.")


# Save output
out_scores_path = os.path.join(DATA_DIR, "phaseC_risk_estimates.csv")
# Include all calculated columns and loan_amnt, default_flag for context
cols_to_save = ['loan_amnt','default_flag','PD_score','LGD_est','LGD_model','outstanding_est','EAD_model','Expected_Loss','RWA']
existing_cols_to_save = [c for c in cols_to_save if c in df.columns]

if existing_cols_to_save:
    df[existing_cols_to_save].head(1000).to_csv(out_scores_path, index=False)
    print(f"‚úÖ Saved first 1000 rows with risk estimates to: {out_scores_path}")
else:
    print("No relevant columns to save for risk estimates.")

# Also save model artifacts list
artifacts = {
    "pd_preprocessor": os.path.join(DATA_DIR, "pd_preprocessor_phaseB.pkl"),
    "pd_model": os.path.join(DATA_DIR, "pd_lgb_phaseB.pkl"),
    "lgd_model": os.path.join(DATA_DIR, "lgd_model.pkl") if os.path.exists(os.path.join(DATA_DIR, "lgd_model.pkl")) else None,
    "ead_model": os.path.join(DATA_DIR, "ead_model.pkl") if os.path.exists(os.path.join(DATA_DIR, "ead_model.pkl")) else None
}
import json
with open(os.path.join(DATA_DIR, "risk_model_artifacts.json"), 'w') as f:
    json.dump(artifacts, f, indent=4)
print("Saved risk model artifact info to:", os.path.join(DATA_DIR, "risk_model_artifacts.json"))
```

```python
# Phase D: Portfolio Visualization & Risk Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Phase C output
df = pd.read_csv("/content/data/phaseC_risk_estimates.csv")

# 1Ô∏è‚É£ Distribution plots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df['PD_score'], bins=50, kde=True, ax=axes[0, 0], color='royalblue') # FIX: Changed 'PD' to 'PD_score'
axes[0, 0].set_title('Probability of Default (PD) Distribution')

# Check if LGD_model exists before plotting
if 'LGD_model' in df.columns:
    sns.histplot(df['LGD_model'], bins=50, kde=True, ax=axes[0, 1], color='tomato') # FIX: Changed 'LGD' to 'LGD_model'
    axes[0, 1].set_title('Loss Given Default (LGD) Distribution')
else:
    axes[0, 1].set_title('LGD Distribution (Data Not Available)') # Handle missing LGD
    axes[0, 1].text(0.5, 0.5, 'LGD model not trained or data missing', horizontalalignment='center', verticalalignment='center', transform=axes[0,1].transAxes)


# Check if EAD_model exists before plotting
if 'EAD_model' in df.columns:
    sns.histplot(df['EAD_model'], bins=50, kde=True, ax=axes[1, 0], color='mediumseagreen') # FIX: Changed 'EAD' to 'EAD_model'
    axes[1, 0].set_title('Exposure at Default (EAD) Distribution')
else:
    axes[1, 0].set_title('EAD Distribution (Data Not Available)') # Handle missing EAD
    axes[1, 0].text(0.5, 0.5, 'EAD model not trained or data missing', horizontalalignment='center', verticalalignment='center', transform=axes[1,0].transAxes)


# Check if Expected_Loss exists before plotting
if 'Expected_Loss' in df.columns:
    sns.histplot(df['Expected_Loss'], bins=50, kde=True, ax=axes[1, 1], color='orange') # FIX: Changed 'EL' to 'Expected_Loss'
    axes[1, 1].set_title('Expected Loss (EL) Distribution')
else:
    axes[1, 1].set_title('Expected Loss (EL) Distribution (Data Not Available)') # Handle missing EL
    axes[1, 1].text(0.5, 0.5, 'Expected Loss not computed', horizontalalignment='center', verticalalignment='center', transform=axes[1,1].transAxes)


plt.tight_layout()
plt.show()

# 2Ô∏è‚É£ Correlation among PD, LGD, EAD, EL, RWA
# Select existing risk columns for correlation
risk_cols = ['PD_score', 'LGD_model', 'EAD_model', 'Expected_Loss', 'RWA']
existing_risk_cols = [c for c in risk_cols if c in df.columns]

if len(existing_risk_cols) >= 2: # Need at least 2 columns to compute correlation
    plt.figure(figsize=(8, 6))
    sns.heatmap(df[existing_risk_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation between Risk Parameters')
    plt.show()
else:
    print("\nSkipping correlation plot: Not enough risk columns available.")


# 3Ô∏è‚É£ Segmentation by loan purpose (if available)
try:
    # Load the original full dataset again to get 'purpose'
    # This assumes the order is the same as the phaseC_risk_estimates.csv
    # A more robust approach would be to merge on a unique identifier if available
    df_full = pd.read_csv("/content/lending_club_loan_two.csv", low_memory=False)

    # Ensure lengths match before combining, and add risk columns
    if len(df_full) >= len(df): # Use >= in case phaseC_risk_estimates was truncated
        df_full_subset = df_full.iloc[:len(df)].copy() # Take the same number of rows
        # Copy the risk columns from df to the subset of df_full
        for col in existing_risk_cols:
             df_full_subset[col] = df[col]

        # Ensure 'purpose' is in the combined dataframe
        if 'purpose' in df_full_subset.columns:
            # Group by purpose and calculate mean PD and EL
            cols_to_group = [col for col in ['PD_score', 'Expected_Loss'] if col in df_full_subset.columns]
            if cols_to_group:
                seg = df_full_subset.groupby('purpose')[cols_to_group].mean().sort_values(cols_to_group[0] if 'PD_score' in cols_to_group else cols_to_group[0], ascending=False).head(10)

                # Plotting - adjust based on available columns
                if 'PD_score' in cols_to_group and 'Expected_Loss' in cols_to_group:
                     seg.plot(kind='bar', figsize=(10, 5), color=['steelblue', 'coral'])
                     plt.title('Top Risk Segments by Loan Purpose (Avg PD & EL)')
                     plt.ylabel('Average Value')
                elif 'PD_score' in cols_to_group:
                     seg['PD_score'].plot(kind='bar', figsize=(10, 5), color='steelblue')
                     plt.title('Top Risk Segments by Loan Purpose (Avg PD)')
                     plt.ylabel('Average PD')
                elif 'Expected_Loss' in cols_to_group:
                     seg['Expected_Loss'].plot(kind='bar', figsize=(10, 5), color='coral')
                     plt.title('Top Risk Segments by Loan Purpose (Avg EL)')
                     plt.ylabel('Average Expected Loss')
                else:
                     print("Skipping segmentation plot: Neither PD_score nor Expected_Loss available.")

                plt.xticks(rotation=45, ha='right')
                plt.tight_layout()
                plt.show()
            else:
                 print("Skipping segmentation plot: Neither PD_score nor Expected_Loss available in combined data.")
        else:
            print("Skipping segmentation plot: 'purpose' column not found in original data.")
    else:
        print(f"Skipping segmentation plot: Length mismatch between loaded data ({len(df)}) and original data ({len(df_full)}).")


except FileNotFoundError:
    print("Segmentation skipped: Original data file not found.")
except Exception as e:
    print(f"Segmentation skipped: {e}")


# 4Ô∏è‚É£ Portfolio summary (already printed in Phase C, but can re-display if needed)
# portfolio_summary = {
#     'Mean_PD': df['PD_score'].mean() if 'PD_score' in df.columns else np.nan,
#     'Mean_LGD': df['LGD_model'].mean() if 'LGD_model' in df.columns else np.nan,
#     'Mean_EAD': df['EAD_model'].mean() if 'EAD_model' in df.columns else np.nan,
#     'Total_Expected_Loss': df['Expected_Loss'].sum() if 'Expected_Loss' in df.columns else np.nan,
#     'Total_RWA': df['RWA'].sum() if 'RWA' in df.columns else np.nan,
# }
# print("üìä Portfolio Risk Summary:")
# for k, v in portfolio_summary.items():
#     print(f"{k:25s}: {v:,.4f}")
```

```python
# ===============================================
# üìò Phase E ‚Äî Final Results Summary + Interpretations
# ===============================================

import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# 1Ô∏è‚É£ Load final risk outputs
df_risk = pd.read_csv("/content/data/phaseC_risk_estimates.csv")

print("\n============================")
print("üìä PORTFOLIO-LEVEL KPIs")
print("============================")
# FIX: Use correct column names from phaseC_risk_estimates.csv
mean_pd = df_risk['PD_score'].mean()
mean_lgd = df_risk['LGD_model'].mean()
mean_ead = df_risk['EAD_model'].mean()
total_el = df_risk['Expected_Loss'].sum()
total_rwa = df_risk['RWA'].sum()

print(f"Average Probability of Default (PD): {mean_pd:.4f}")
print(f"Average Loss Given Default (LGD):    {mean_lgd:.4f}")
print(f"Average Exposure at Default (EAD):   {mean_ead:.4f}")
print(f"Total Expected Loss (EL):            {total_el:,.2f}")
print(f"Total Risk-Weighted Assets (RWA):    {total_rwa:,.2f}")

print("\nüß† INTERPRETATION:")
print(f"- The mean PD of {mean_pd:.2%} indicates the average probability that a borrower defaults.")
print(f"- LGD ‚âà {mean_lgd:.2%} means that, on average, {mean_lgd:.2%} of the exposure is lost when default occurs.")
print("- Average EAD shows the typical exposure per borrower at the time of default.")
print("- Expected Loss (EL = PD √ó LGD √ó EAD) quantifies the overall portfolio loss expectation.")
print("- RWA converts credit risk into capital requirement under Basel III.\n")

# 2Ô∏è‚É£ Feature importance from LightGBM model
print("================================")
print("üìà TOP FEATURE IMPORTANCES")
print("================================")
try:
    # Load the LightGBM model saved in Phase B
    # We saved the sklearn estimator in Phase B as pd_lgb_phaseB.pkl
    model_lgbm = joblib.load("/content/data/pd_lgb_phaseB.pkl")

    # Try to retrieve feature names from the preprocessor used in Phase B
    preproc_b = joblib.load("/content/data/pd_preprocessor_phaseB.pkl")
    try:
        num_names = preproc_b.transformers_[0][2] # Numeric feature names
        ohe = preproc_b.named_transformers_['cat'].named_steps['ohe']
        cat_names = list(ohe.get_feature_names_out(preproc_b.transformers_[1][2])) # Categorical feature names
        feature_names_b = num_names + cat_names
    except Exception as e:
        print("Could not retrieve feature names from Phase B preprocessor:", e)
        # Fallback: Use feature names from the LightGBM model if available (might be less descriptive)
        if hasattr(model_lgbm, 'feature_name_'):
            feature_names_b = model_lgbm.feature_name_
        else:
            feature_names_b = [f"feature_{i}" for i in range(model_lgbm.n_features_in_)]


    # Use feature importances from the trained sklearn estimator
    importances = model_lgbm.feature_importances_ # Use feature_importances_ attribute for LGBMClassifier

    imp_df = pd.DataFrame({'feature': feature_names_b, 'importance': importances})
    # Use 'importance' as sort key for feature_importances_ (gain was for booster)
    imp_df = imp_df.sort_values('importance', ascending=False).head(15)


    print(imp_df)
    plt.figure(figsize=(8,5))
    # Use 'importance' for x-axis
    sns.barplot(x='importance', y='feature', data=imp_df, palette='viridis')
    plt.title("Top 15 Important Features (LightGBM)") # Adjusted title
    plt.xlabel("Importance") # Adjusted label
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

    print("\nüß† INTERPRETATION:")
    print("- Features with highest importance (e.g., interest rate, loan-to-income, DTI) drive default risk.")
    print("- High int_rate or high loan_to_income ‚Üí greater PD likelihood.")
    print("- These features should be monitored in credit policy and underwriting.\n")
except FileNotFoundError:
     print("Feature importance unavailable: pd_lgb_phaseB.pkl or pd_preprocessor_phaseB.pkl not found.")
except Exception as e:
    print(f"Feature importance unavailable: {e}")

# 3Ô∏è‚É£ PD Distribution
# FIX: Use correct column name 'PD_score'
if 'PD_score' in df_risk.columns:
    sns.histplot(df_risk['PD_score'], bins=50, kde=True, color='royalblue')
    plt.title("Distribution of Predicted Probability of Default (PD)")
    plt.xlabel("PD")
    plt.ylabel("Frequency")
    plt.show()
    print("\nüß† INTERPRETATION:")
    print("- PD distribution shows most borrowers have low default probability, forming a right-skewed curve.")
    print("- A small tail with higher PD values represents high-risk borrowers who need closer monitoring.\n")
else:
    print("Skipping PD distribution plot: 'PD_score' column not found in df_risk.")


# 4Ô∏è‚É£ Risk segmentation by loan purpose
print("=========================================")
print("üè¶ RISK SEGMENTATION BY LOAN PURPOSE")
print("=========================================")
try:
    # Load the original full dataset again to get 'purpose'
    df_raw = pd.read_csv("/content/lending_club_loan_two.csv", low_memory=False)
    # Merge risk data to raw data, aligning by index (assuming same row order)
    # Make sure both dataframes have a unique index to merge on, or reset index
    # A more robust way would be to merge on a common identifier column if available
    # For now, assume index alignment after loading
    df_raw_indexed = df_raw.reset_index()
    df_risk_indexed = df_risk.reset_index()

    # Merge based on the reset index
    merged_df = pd.merge(df_raw_indexed, df_risk_indexed[['index','PD_score', 'Expected_Loss']], on='index', how='left') # FIX: Use correct column names
    merged_df = merged_df.drop(columns='index') # Drop the merge index

    # Ensure 'purpose' is in the merged dataframe and risk columns were merged successfully
    if 'purpose' in merged_df.columns and 'PD_score' in merged_df.columns and 'Expected_Loss' in merged_df.columns: # FIX: Use correct column names
        # Group by purpose and calculate mean PD and EL
        seg = merged_df.groupby('purpose')[['PD_score','Expected_Loss']].mean().sort_values('Expected_Loss', ascending=False).head(10) # FIX: Use correct column names
        print(seg)

        seg.plot(kind='bar', figsize=(10,5), color=['royalblue','tomato'])
        plt.title("Average PD & Expected Loss by Loan Purpose")
        plt.ylabel("Average Value")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        print("\nüß† INTERPRETATION:")
        print("- Loan purposes with higher EL (e.g., small business or renewable energy) carry greater credit risk.")
        print("- These segments could require stricter credit limits or higher interest rates to compensate for risk.\n")
    else:
        print("Skipping segmentation plot: 'purpose' column missing or risk columns not merged successfully.")


except FileNotFoundError:
    print("Segmentation skipped: Original data file not found.")
except Exception as e:
    print(f"Segmentation skipped: {e}")


# 5Ô∏è‚É£ Correlation among PD, LGD, EAD, EL, RWA
# FIX: Use correct column names for correlation matrix
correlation_cols = ['PD_score','LGD_model','EAD_model','Expected_Loss','RWA']
existing_correlation_cols = [c for c in correlation_cols if c in df_risk.columns]

if len(existing_correlation_cols) >= 2:
    plt.figure(figsize=(6,5))
    sns.heatmap(df_risk[existing_correlation_cols].corr(),
                annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation between Risk Metrics")
    plt.tight_layout()
    plt.show()

    print("\nüß† INTERPRETATION:")
    print("- PD and EL are strongly correlated, as expected (EL = PD √ó LGD √ó EAD).")
    print("- RWA correlates positively with PD and EAD, confirming capital charges grow with exposure and risk.")
    print("- LGD shows weaker correlation since it depends on collateral and recovery policies.\n")
else:
     print("Skipping correlation plot: Not enough risk columns available in df_risk.")


# 6Ô∏è‚É£ Final summary table
# FIX: Use correct column names for final summary table
final_summary = pd.DataFrame({
    "Metric": ["Mean PD","Mean LGD","Mean EAD","Total Expected Loss","Total RWA"],
    "Value": [mean_pd, mean_lgd, mean_ead, total_el, total_rwa]
})
print("=========================================")
print("üìã FINAL PORTFOLIO SUMMARY")
print("=========================================")
print(final_summary)

final_summary.to_csv("/content/data/final_portfolio_summary.csv", index=False)
print("‚úÖ Saved final summary ‚Üí /content/data/final_portfolio_summary.csv")

print("\nüß† INTERPRETATION:")
print("- The summary provides a concise snapshot of the bank‚Äôs credit risk profile.")
print("- These values feed into Expected Credit Loss (ECL) and Basel capital planning.")
print("- In practice, risk officers use them for stress-testing and portfolio optimization.\n")
```
